```python
import pandas as pd

url_dados = 'https://github.com/AsonCS/DataScienceTest/raw/main/quina_data.csv'

# dados['Mutiplos Ganhadores'] = withCountry.apply(lambda city : 1 if len(city.split(',')) > 1 else 0)
# dados['Mutiplos Ganhadores'] = dados['Mutiplos Ganhadores'].apply(lambda val : 0 if val == 'NaN' else val)
# dados['Mutiplos Ganhadores'][dados['Mutiplos Ganhadores'] == NaN]
dados = pd.read_csv(url_dados, sep=';')
dados['cidade'] = dados['cidade'].fillna('')

# dados['acumulado'].unique()
# # # Output: array(['NAO', 'SIM'], dtype=object)
dados['acumulado'] = dados['acumulado'].apply(lambda val : 1 if val == 'SIM' else 0)

dados
```

```python
# def mutipleCities (city):
#   if len(city) == 0:
#     return 0
#   else:
#     if len(str(city).split(',')) > 1:
#       return 1
#     else:
#       return 0

dados['mutiplos_ganhadores'] = dados['cidade'].apply(lambda city : 1 if len(city.split(',')) > 1 else 0)
dados[['ganhadores_quina', 'cidade', 'mutiplos_ganhadores']].iloc[2061:2069]
```

```python
dados['tem_ganhadores'] = dados['cidade'].apply(lambda city : 0 if len(city) == 0 else 1)
dados[['ganhadores_quina', 'cidade', 'mutiplos_ganhadores', 'tem_ganhadores']].iloc[2061:2069]
```
